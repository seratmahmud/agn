```python
file_path = 'ascii/J14301603+2303445_2023-01-06_COS.dat'

t = Table.read(file_path, format='ascii', delimiter='\s', header_start=3, data_start=5)

column_names = ['wl', 'flux', 'flux_err']

original_colnames = t.colnames

for original, new in zip(original_colnames, column_names):
    t.rename_column(original, new)


```

```python


# Plotting

lyman_alpha = 1215

plt.figure(figsize=(10, 6))

plt.plot(t["wl"], t["flux"], label='Spectrum', color="black")

plt.fill_between(t["wl"], t["flux"] - t["flux_err"], t["flux"] + t["flux_err"], color='gray', alpha=0.5, label='Error')

plt.axvline(x=lyman_alpha, color='r', linestyle='--', linewidth=1, label="Lyman alpha")


plt.xlabel("Wavelength (A)")
plt.ylabel("Flux (erg/s/cm^2/A)")

plt.xlim(1100, 1650)
plt.ylim(-100, 400)

plt.legend()

plt.show()
```

```python
z = 0.081

#t["wl"] = t["wl"] / (1 + z)

lyman_alpha = 1215


bin_width = 3

bins = np.arange(min(t["wl"]), max(t["wl"]) + bin_width, bin_width)
bin_centers = 0.5 * (bins[:-1] + bins[1:])

indices = np.digitize(t["wl"], bins)

binned_wl = []
binned_flux = []
binned_flux_err = []


for i in range(1, len(bins)):
    in_bin = indices == i
    if np.any(in_bin):
        binned_wl.append(np.mean(t["wl"][in_bin]))
        binned_flux.append(np.mean(t["flux"][in_bin]))
        binned_flux_err.append(np.mean(t["flux_err"][in_bin]))
    else:
        pass

binned_wl = np.array(binned_wl)
binned_flux = np.array(binned_flux)
binned_flux_err = np.array(binned_flux_err)

plt.figure(figsize=(10, 6))
plt.plot(binned_wl, binned_flux, label='Binned Spectrum', color="black", drawstyle='steps-mid')
plt.fill_between(binned_wl, binned_flux - binned_flux_err, binned_flux + binned_flux_err, color='gray', alpha=0.5, label='Error')
plt.axvline(x=lyman_alpha, color='r', linestyle='--', linewidth=1, label="Lyman alpha")
plt.xlabel("Wavelength ($\AA$)")
plt.ylabel("$f_{\lambda} 10^{-17} erg s^{-1} cm^{-2} \AA^{-1}$")
plt.xlim(1000, 1650)
plt.ylim(-100, 400)
plt.legend()
plt.show()
```

```python


t = Table.read('HET/combined_spectrum_20230113_0000021_exp01.fits')





wavelength = t['Wavelength']
flux = t['F_lambda']
flux_err = t['e_F_lambda']


finite_mask = np.isfinite(flux) & np.isfinite(flux_err) & (flux_err > 0)
wavelength = wavelength[finite_mask]
flux = flux[finite_mask]
flux_err = flux_err[finite_mask]



plt.figure(figsize=(10, 6))
plt.errorbar(wavelength, flux, yerr=flux_err, fmt='.', label='Observed Spectrum')
plt.xlabel('Wavelength')
plt.ylabel('Flux')
plt.title('AGN Spectrum')
plt.legend()
plt.show()

# Model ???
power_law = models.PowerLaw1D(amplitude=1, x_0=1, alpha=-2)
gaussian = models.Gaussian1D(amplitude=1, mean=4861, stddev=10)

combined_model = power_law + gaussian


fitter = fitting.LevMarLSQFitter()
fitted_model = fitter(combined_model, wavelength, flux, weights=1/flux_err,  maxiter=1000)



fit_info = fitter.fit_info
print("Fitting Status:", fit_info['message'])

if fit_info['ierr'] not in [1, 2, 3, 4]:
    print("Warning: The fit may not have converged.")
    print("Details:", fit_info)




#Plotting the fitted model
plt.figure(figsize=(10, 6))
plt.errorbar(wavelength, flux, yerr=flux_err, fmt='.', label='Observed Spectrum')
plt.plot(wavelength, fitted_model(wavelength), label='Fitted Model', color='red')
plt.xlabel('Wavelength')
plt.ylabel('Flux')
plt.title('Fitted AGN Spectrum')
plt.legend()
plt.show()

#Extracting and plotting the quasar continuum
continuum = fitted_model[0](wavelength)  # Assuming the first component is the continuum
plt.figure(figsize=(10, 6))
plt.plot(wavelength, continuum, label='Quasar Continuum', color='green')
plt.xlabel('Wavelength')
plt.ylabel('Flux')
plt.title('Quasar Continuum')
plt.legend()
plt.show()

#Subtracting the quasar continuum from the original flux
residual_flux = flux - continuum

#Plotting the spectrum after subtracting the quasar continuum
plt.figure(figsize=(10, 6))
plt.errorbar(wavelength, residual_flux, yerr=flux_err, fmt='.', label='Spectrum without Continuum')
plt.xlabel('Wavelength')
plt.ylabel('Flux')
plt.title('AGN Spectrum Without Quasar Continuum')
plt.legend()
plt.show()
```

```python
table = Table.read('HET/combined_spectrum_20230113_0000021_exp01.fits')
wavelength = table['Wavelength']
flux = table['F_lambda']


dust_map = fits.open('sfddata-master/SFD_dust_4096_ngp.fits')
ebv_data = dust_map[0].data


coords = SkyCoord('14h30m16.0414s', '+23d03m44.53s', frame='icrs')

wcs_info = WCS(dust_map[0].header)
pixel_coords = wcs_info.world_to_pixel(coords)

ebv_value = ebv_data[int(pixel_coords[1]), int(pixel_coords[0])]

dust_map.close()


R_V = 3.1
A_V = R_V * ebv_value

wavelength_microns = wavelength / 1e4

extinction_curve = fitzpatrick99(wavelength_microns, A_V, R_V)
extinction_curve = np.minimum(extinction_curve, 10)

corrected_flux = flux * 10**(0.4 * extinction_curve)
print(extinction_curve)
print(corrected_flux)

finite_mask = np.isfinite(corrected_flux) & np.isfinite(flux_err) & (flux_err > 0)
wavelength = table['Wavelength'] 
wavelength = wavelength[finite_mask]
corrected_flux = corrected_flux[finite_mask]
flux_err = flux_err[finite_mask]


plt.figure(figsize=(10, 6))
plt.errorbar(wavelength, corrected_flux, yerr=flux_err, fmt='.', label='Spectrum without Continuum')
plt.xlabel('Wavelength')
plt.ylabel('Flux')
plt.title('DUST corrected flux')
plt.legend()
plt.show()
```

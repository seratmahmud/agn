```python
from astropy.io import fits
from astropy.table import Table, vstack, hstack
import matplotlib.pyplot as plt
import numpy as np

from scipy.optimize import curve_fit
```

# Data

```python
#file_name = 'bundle_20230529_standards/combined_spectrum_20230112_0000009_exp01.fits'
file_name = 'bundle_20230529_standards/spectrum_20230112_0000010_exp01_red.fits'
hdul = fits.open(file_name)
data = hdul[0].data
hdr = hdul[0].header

z = 0.081


def model_func(x, a, b):
    return a*x + b


wl = data[0]
wl_o = wl/(1+z)
flux_o = data[1] *(10**16)
err_o = data[2]

mask0 = np.where((wl_o>6000) & (wl_o < 7000)) 
wl = wl_o[mask0]
flux = flux_o[mask0]
err = err_o[mask0]

mask1 = np.where((wl < 6015) | (wl > 6170)) 
wl = wl[mask1]
flux = flux[mask1]
err = err[mask1]

mask2 = np.where((wl < 6340) | (wl > 6480)) 
wl = wl[mask2]
flux = flux[mask2]
err = err[mask2]

mask3 = np.where((wl < 6610) | (wl > 6800)) 
wl = wl[mask3]
flux = flux[mask3]
err = err[mask3]


params, covariance = curve_fit(model_func, wl, flux)
a_fit, b_fit = params
wl_fit = np.linspace(wl.min(), wl.max(), len(wl_o)) 
flux_fit = model_func(wl_fit, a_fit, b_fit)

plt.figure(figsize=(10, 6))
plt.scatter(wl, flux, label='Data')  # Original data
plt.plot(wl_fit, flux_fit-flux_o, label='Best fit curve')  # Fitted curve
plt.xlabel('Wavelength (wl)')
plt.ylabel('Flux')
plt.legend()
plt.show()
```
